# **1. Importing the Data & libraries**

```python
# Import the libaries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('/content/TWO_CENTURIES_OF_UM_RACES.csv', on_bad_lines='skip')     # Skip problematic lines

# Display the first few rows
df.head(10)
```


# **2. Data Preprocessing**

##***2.1 Check Data***##

```python
# Check data types and null values
print(df.info())
print(df.isna().sum())

# Display summary statistics
print(df.describe())
```

##***2.2 Clean up the Data*** ##

### ***2.2.1 USA & 50Km-50mi & 2018***

```python
# First Step: Filter rows where the 'Event distance/length' is either '50km' or '50mi'
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
# Second Step: Filter rows for specific event distances ('50km', '50mi') and the year of event (2018)
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2018)]
```

```python
# Third Step :Filter rows where the event name contains 'USA' in the country part
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# Combine filters for both 'Event name' and other conditions
df2 = df[
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') &
    (df['Event distance/length'].isin(['50km', '50mi'])) &
    (df['Year of event'] == 2018)
]
df2
```

### ***2.2.2 Remove USA from Event name***

```python
# Remove content within parentheses, including the parentheses, from 'Event name'
df2['Event name'] = df2['Event name'].str.split('(').str.get(0).str.strip()
df2
```

### ***2.2.3 Remove 'h' from Performance***

```python
# Remove 'h' or extra characters from 'Athlete performance'
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2
```

## ***2.3 Athlete Age column***

```python
# Calculate athlete age for the year 2018
df2['athlete_age'] = 2018 - df2['Athlete year of birth']
df2
```

## ***2.4 Drop Columns***

```python
# Drop specified columns
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df2
```

## ***2.5 Check and Delete Null Values***

```python
# Check for null values in each column
df2.isna().sum()
```

```python
# Filter rows where 'athlete_age' has null values
df2[df2['athlete_age'].isna() == True]
```

```python
# Delete Null values from athlete_age column
df2 = df2.dropna(subset=['athlete_age'])
df2
```

## ***2.6 Check Duplicates***

```python
# Check for duplicate rows
df2[df2.duplicated() == True]
```

## ***2.7 Reset Index***

```python
# Reset the index of the DataFrame and drop the old index
df2 = df2.reset_index(drop=True)
df2
```

## ***2.8 Rename the Columns***

```python
# Rename columns in the DataFrame
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    'athlete_age': 'athlete_age'
})
df2
```

## ***2.9 Reorder the Columns***

```python
# Drop 'Year of event' and reorder columns
df2 = df2.drop(columns=['Year of event'], errors='ignore')  # Remove the column, ignoring errors if it doesn't exist

# Reorder columns as needed
df2 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers',
           'athlete_performance', 'athlete_gender', 'athlete_average_speed', 'athlete_id','athlete_age']]
df2
```

## ***2.10 Exclude X Gender***

```python
# Exclude rows where 'athlete_gender' is 'X'
df2 = df2[df2['athlete_gender'] != 'X']
```

## ***2.11 Race Season Column***

```python
# Extract the month from 'race_day' column
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
# Assign seasons based on the race_month column
df2['race_season'] = df2['race_month'].apply(
    lambda x: 'Winter' if x > 11 or x < 3 else
              'Fall' if x > 8 else
              'Summer' if x > 5 else
              'Spring'
)
df2
```

## ***2.12 Convert Data Types***

```python
df2['athlete_average_speed'] = pd.to_numeric(df2['athlete_average_speed'], errors='coerce')
df2 = df2.dropna(subset=['athlete_average_speed'])
```

# **3. EDA**

## ***3.1 Distribution of Race Length***

```python
# Create a histogram for 'race_length'

sns.histplot(df2['race_length'])
plt.xlabel('Race Length')
plt.ylabel('Frequency')
plt.title('Distribution of Race Length')
plt.show()
```

## ***3.2 Distribution of Race Length by Gender***

```python
# Create a histogram for 'race_length' with gender distinction

sns.histplot(data=df2, x='race_length', hue='athlete_gender', multiple='stack')
plt.xlabel('Race Length')
plt.ylabel('Frequency')
plt.title('Distribution of Race Length by Gender')
plt.show()
```

## ***3.3 Distribution of Athlete Average Speed***

```python
# Create a distribution plot for 'athlete_average_speed' for 50mi race

sns.histplot(data=df2, x='athlete_average_speed', bins=30, kde=False)
plt.xlabel('Athlete Average Speed')
plt.ylabel('Count of People')
plt.title('Distribution of Athlete Average Speed')
plt.show()

```

## ***3.4 Distribution of Athlete Average Speed by Race Length and Gender***

```python
# Create a violin plot
sns.violinplot(data=df2, x='race_length', y='athlete_average_speed', hue='athlete_gender', split=True)
plt.xlabel('Race Length')
plt.ylabel('Athlete Average Speed')
plt.title('Distribution of Athlete Average Speed by Race Length and Gender')
plt.show()
```

## ***3.5 Relationship Between Athlete Age and Average Speed by Gender***

```python

# Create a scatterplot with regression lines
sns.lmplot(data=df2, x='athlete_age', y='athlete_average_speed', hue='athlete_gender', height=6, aspect=1.5)
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Average Speed')
plt.title('Relationship Between Athlete Age and Average Speed by Gender')
plt.show()
```

## ***3.6  mean average speed by race length and gender***

```python
# Calculate mean average speed grouped by race length and gender
grouped_data = df2.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()

# Display the results for 50k and 50mi
print(grouped_data.loc[['50km', '50mi']])
```

## ***3.7 Top & Worst 10-performing age groups***

```python
# Filter for 50mi race, group by athlete age, and calculate mean speed and count
result = (
    df2.query('race_length == "50mi"')
    .groupby('athlete_age')['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending=False)
    .query('count > 19')
    .head(10)
)

# Display the result
print(result)
```

```python
# Filter for 50mi race, group by athlete age, and calculate mean speed and count
result = (
    df2.query('race_length == "50mi"')
    .groupby('athlete_age')['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending=True)
    .query('count > 9')
    .head(20)  # Display the top 20 worst-performing age groups
)

# Display the result
print(result)
```

## ***3.8 Seasonal_analysis***

```python
# Group by race_season, calculate mean and count, and sort by mean in descending order
seasonal_analysis = (
    df2.groupby('race_season')['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending=False)
)

# Display the results
print(seasonal_analysis)
```

```python
# Filter for 50mi races, group by race season, calculate mean and count, and sort by mean
seasonal_50mi_analysis = (
    df2.query('race_length == "50mi"')
    .groupby('race_season')['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending=False)
)

# Display the results
print(seasonal_50mi_analysis)
```
